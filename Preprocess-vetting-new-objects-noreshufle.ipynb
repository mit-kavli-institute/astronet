```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
## Load the two tables, the original training set and the new objects
fname = 'vetting-v01'
fname_new_TOIs = 'vetting-new-events-withfilenames'

fpath = f'mnt/tess/labels/{fname}.csv'
fpath_new_TOIs=f'mnt/tess/labels/{fname_new_TOIs}.csv'

# fpath = f'mnt/tess/labels/{fname}.csv'
# all_table = pd.read_csv(fpath, header=0, low_memory=False).set_index('Astro ID')
# long_per_table= pd.read_csv(fpath_long_periods,header=0,low_memory=False).set_index('Astro ID')

# Read CSVs and set 'Astro ID' as index
all_table = pd.read_csv(fpath, index_col='Astro ID', low_memory=False)
print('Original table shape:', all_table.shape)
new_table = pd.read_csv(fpath_new_TOIs, index_col='Astro ID', low_memory=False)
print('New events table shape:', new_table.shape)


# all_table_2=pd.concat([all_table, new_table], ignore_index=False)

all_table=all_table
print('Final table shape:', all_table.shape)

# (Optional) Check for duplicate indices
if all_table.index.duplicated().any():
    print("Warning: There are duplicate Astro ID indices.")
```

```python
#Treat the old and new events in parallel

all_table = all_table.rename(columns={'filename': 'File',
                                      'Period': 'Per',
                                      'Duration': 'Dur',
                                      'Transit_Depth': 'Depth',
                                      'star_rad': 'SRad',
                                      'star_rad_est': 'SRadEst',
                                      'star_mass': 'SMass',})

new_table = new_table.rename(columns={'filename': 'File',
                                      'Period': 'Per',
                                      'Duration': 'Dur',
                                      'Transit_Depth': 'Depth',
                                      'star_rad': 'SRad',
                                      'star_rad_est': 'SRadEst',
                                      'star_mass': 'SMass',})
```

```python
## Set labels
def set_labels(row):
    a = ~row.isna()

    if isinstance(row["Final"], str) and len(row["Final"]) == 1:
        return row  # Skip rows where "Final" is a single character


    if row['Final'] == 'i':
        # skip objects labeled as "inside the star"
        return row

    if a['Final']:
        print("Row Final:", row["Final"], type(row["Final"]))

        row[f'disp_{row["Final"][0]}'] = 1
        row[f'disp_{row["Final"][1]}'] = 1
    else:
        for user in users:
            if a[user] and row[user]:
                row[f'disp_{row[user][0]}'] += 1
                row[f'disp_{row[user][1]}'] += 1

    return row

all_table = all_table.apply(set_labels, axis=1)

new_table = new_table.apply(set_labels, axis=1)
```

```python
# Ensure columns match before splitting
common_columns = all_table.columns.intersection(new_table.columns)
new_table = new_table[common_columns]

# 1. Recreate the original split for the old dataset
t_train_orig, t_test_orig = train_test_split(all_table, test_size=0.1, random_state=42)
t_train_orig, t_val_orig = train_test_split(t_train_orig, test_size=1./9, random_state=42)

# 2. Split the new dataset separately
t_train_new, t_test_new = train_test_split(new_table, test_size=0.1, random_state=42)
t_train_new, t_val_new = train_test_split(t_train_new, test_size=1./9, random_state=42)

# 3. Merge train, val, and test sets separately
t_train = pd.concat([t_train_orig, t_train_new]).sample(frac=1, random_state=42)
t_val = pd.concat([t_val_orig, t_val_new]).sample(frac=1, random_state=42)
t_test = pd.concat([t_test_orig, t_test_new]).sample(frac=1, random_state=42)

# Print sizes and duplicates
print(f'Split sizes. Train: {len(t_train)}; Valid: {len(t_val)}; Test: {len(t_test)}')
print(f'Duplicate TICs: {len(all_table.index.values) - len(set(all_table.index.values))}')
print('Splits')
print('  train:', len(t_train))
print('  val:', len(t_val))
print('  test:', len(t_test))

# Check label arrays to ensure no row is missing labels
assert not any((t_train[['disp_e', 'disp_p', 'disp_n', 'disp_b', 'disp_t', 'disp_u', 'disp_j']].sum(axis=1)) == 0)
assert not any((t_val[['disp_e', 'disp_p', 'disp_n', 'disp_b', 'disp_t', 'disp_u', 'disp_j']].sum(axis=1)) == 0)
assert not any((t_test[['disp_e', 'disp_p', 'disp_n', 'disp_b', 'disp_t', 'disp_u', 'disp_j']].sum(axis=1)) == 0)

# File Naming
NAME = "tois-triageJs-nocentroid-newTOIs2025-v3-noreshufle"

# Save only if files do not already exist
save_path = f'mnt/tess/astronet/tces-{fname}-{NAME}-all.csv'
if os.path.exists(save_path):
    print("*****\nAlready saved TCE csv files. Are you sure you want to overwrite them?\n*****")
else:
    t_train.to_csv(f'mnt/tess/astronet/tces-{fname}-{NAME}-train.csv')
    t_val.to_csv(f'mnt/tess/astronet/tces-{fname}-{NAME}-val.csv')
    t_test.to_csv(f'mnt/tess/astronet/tces-{fname}-{NAME}-test.csv')
    all_table.to_csv(f'mnt/tess/astronet/tces-{fname}-{NAME}-all.csv')

print("Dataset splits successfully saved.")
```
